```python
 # Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable and the data into a Pandas DataFrame
csv_path = Path("../archive (1)/user_behavior_dataset.csv")
user_behavior_df = pd.read_csv(csv_path)
user_behavior_df.head()
```

```python
# Age Min and Max
 # Calculate the earliest/latest year a book was published
youngest = user_behavior_df["Age"].min()
oldest = user_behavior_df["Age"].max()

# Print the earliest and latest year of publication.
print(youngest)
print(oldest)
```

```python
# Create the bins in which Data will be held
# Bins are 0, 59.9, 69.9, 79.9, 89.9, 100.   
bins = [18, 28, 38, 48, 59]

# Create the names for the five bins
group_names = ["Young Adults", "Late 20- 30's", "Late 30-40's", "Late 40-50's"]
```

```python
# Filter to only buildings with infested units greater than 0
# new DF for column
user_behavior_df["cell_phone_age_cohort"]=pd.cut(user_behavior_df["Age"],bins, labels=group_names)
user_behavior_df.head()
```

```python
# Average User Behavior Class by age
cell_age_usage_group = user_behavior_df.groupby("cell_phone_age_cohort")
new_df=cell_age_usage_group[["App Usage Time (min/day)", "Screen On Time (hours/day)","Battery Drain (mAh/day)", "Number of Apps Installed", "Data Usage (MB/day)" ]].mean()
new_df.head()
```

```python
# Sum of  User Behavior Class by age
columns_to_scale = ['App Usage Time (min/day)']
new_df[columns_to_scale] = new_df[columns_to_scale] 
new_df.applymap(lambda x: f"{x / 100:.02%}").head()
```

```python
# Average User Behavior Class by age
device_usage_group = user_behavior_df.groupby("Device Model")
device_model_df=device_usage_group[["App Usage Time (min/day)", "Screen On Time (hours/day)","Battery Drain (mAh/day)", "Number of Apps Installed", "Data Usage (MB/day)" ]].mean()
device_model_df.head()
```
